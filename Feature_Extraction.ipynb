```python
# Features:
# Vocabulary lists by level (A1, A2, B1, B2, C1, C2) (C-level not really used for learners)
# Frequency list
# POS
# Flesch reading ease - Christina
# Syllabification
# average sentence length
# type-to-token ratio
# POS proportions
# Number of tokens and types
# number of sentences in the text
# other syntactic information
```

```python
# Vocab lists by level
# Read in the vocab list
vocab_lists = {'A1': [],
               'A2': [],
               'B1': [],
               'B2': []}
with open('/content/drive/MyDrive/capstone/Vocab/spanish1.txt') as sp1_file:
  A1_list = sp1_file.read().split('\n')
  vocab_lists['A1'] = A1_list

with open('/content/drive/MyDrive/capstone/Vocab/spanish2.txt') as sp2_file:
  A2_list = sp2_file.read().split('\n')
  vocab_lists['A2'] = A2_list
```

```python
# Example text
text = "Hola me llamo Maria y tengo dieciséis años".split(" ")
```

```python

word_counts = {'A1': 0, 
               'A2': 0,
               'B1': 0,
               'B2': 0, 
               'OOV': 0}
# Find individual words
for word in text:
  in_vocab_flag = False
  for level in vocab_lists.keys():
    if word in vocab_lists[level]:
      word_counts[level] += 1
      in_vocab_flag = True
  if not in_vocab_flag:
    word_counts['OOV'] += 1

# TODO - match n-grams

# Find percentage of words in text by level
CUTOFF = 0.85
text_len = len(text)
if word_counts['A1']/text_len >= CUTOFF:
  text_level = 'A1'
elif word_counts['A1'] + word_counts['A2'] >= CUTOFF:
  text_level = 'A2'
elif word_counts['A1'] + word_counts['A2'] + word_counts['B1'] >= CUTOFF:
  text_level = 'B1'
elif word_counts['A1'] + word_counts['A2'] + word_counts['B1'] + word_counts['B2'] >= CUTOFF:
  text_level = 'B2'
else:
  text_level = 'C'

print(word_counts)
print(text_level)
```
