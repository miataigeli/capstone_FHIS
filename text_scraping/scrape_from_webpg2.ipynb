```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
###################################### A1-level texts ###################################
#### stories stored as string in a1_str_story_list and a2_str_story_list
#### 23 A1 stories and 33 A2 stories 

NUM_A1 = 23
NUM_A2 = 33

wb_address = 'https://www.gutenberg.org/files/15353/15353-h/15353-h.htm'
f = urlopen(wb_address)
html_soup = BeautifulSoup(f, 'lxml')
a1_story_list = []
node = html_soup.find('a', id = 'e1')
for i in range(NUM_A1):
    next_node = html_soup.find('a', id = 'e' + str(i+2))
    story = u""
    for tag in node.next_siblings:
        if tag == next_node:
            a1_story_list.append(story)
            node = next_node
            break
        else:
            story += str(tag)
            
a2_story_list = []
node = html_soup.find('a', id='e24')
for i in range(NUM_A2):
    next_node = html_soup.find('a', id = 'e' + str(i+25))
    story = u""
    for tag in node.next_siblings:
        if tag == next_node:
            a2_story_list.append(story)
            node = next_node
            break
        else:
            story += str(tag)
```

```python
### Store A1 stories as string #####
a1_str_story_list = []
for i in range(NUM_A1):
    story_soup = BeautifulSoup(a1_story_list[i], 'lxml')
    str_story = story_soup.get_text()
    a1_str_story_list.append(str_story)
```

```python
### Store A2 stories as string #####
a2_str_story_list = []
for i in range(NUM_A2):
    story_soup = BeautifulSoup(a2_story_list[i], 'lxml')
    str_story = story_soup.get_text()
    a2_str_story_list.append(str_story)
```
