```python
import os
from tika import parser
```

```python
pdf_dir = '/Users/eun-youngchristinapark/Documents/UBC-CAPSTONE/pdf_scraping/'
wb_address = 'https://lingua.com/spanish/reading/#exercises'
```

```python
pdf_files = [filename for filename in os.listdir(pdf_dir)]
pdf_files
```

```python
def extract_text(dir_name):
    '''
    This function extracts the content of pdf files.
    ------------------------------------------
    Argument: dir_name: (str) name of the directory which contains the pdf files
    returns texts: (list) list of strings where each string is the content of the pdf file.
    '''
    texts = []
    level_list = []
    pdf_files = [filename for filename in os.listdir(dir_name)]
    for file in pdf_files:
        file_name = dir_name + file
        if file != '.DS_Store':
            raw = parser.from_file(file_name)
            text_content = raw['content'].strip()
            text_list = text_content.split('\n\n')
            ts = ''
            for text_string in text_list:
                if text_string.startswith('¿'):
                    break
                else:
                    ts = ts + text_string
        
            texts.append(ts)
            level_list.append(file[:2])
    return texts, level_list
```

```python
a1a2_texts, level_list = extract_text(pdf_dir)
#print(len(a1a2_texts))
for i, ts in enumerate(a1a2_texts):
    print(f'{i+1} \n{ts} \n')
```

```python
# Creating a json file

import json

SOURCE = wb_address
AUTHOR = 'Unknown'
TITLE = 'Not Available'
LEVEL_A1 = 'A1'
LEVEL_A2 = 'A2'
dict_list = []

for i in range(len(a1a2_texts)):
    story_data = {}
    story_data['source'] = SOURCE
    story_data['author'] = AUTHOR
    story_data['title'] = TITLE
    story_data['level'] = level_list[i].title()
    story_data['content'] = a1a2_texts[i]
    dict_list.append(story_data)
        
with open('pdf_a1_a2.txt', 'w') as outfile:
        json.dump(dict_list, outfile)
```
