```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
###################################### B1-level texts ###################################
#### stories stored as string in str_story_list
####

wb_address = 'https://www.gutenberg.org/files/36805/36805-h/36805-h.htm'
f = urlopen(wb_address)
html_soup = BeautifulSoup(f, 'lxml')
story_list = []
node = html_soup.find('h3')
next_node = node
while next_node != None:
    next_node = None
    story = u""
    for tag in node.next_siblings:
        if tag.name == 'h3':
            story_list.append(story)
            node = tag
            next_node = tag
            break
        else:
            story += str(tag)
```

```python
str_story_list = []
for i in range(3,35):  # exclude 0, 1, 2, and 21
    if i != 21:
        story_soup = BeautifulSoup(story_list[i], 'lxml')
        str_story = story_soup.get_text()
        str_story_list.append(str_story)
```

```python
# Creating a json file

import json

SOURCE = wb_address
AUTHOR = 'Elijah Clarence Hills'
TITLE = 'Spanish Tales for Beginners'
LEVEL = 'B1'
dict_list = []

for i in range(len(str_story_list)):
    story_data = {}
    story_data['source'] = SOURCE
    story_data['author'] = AUTHOR
    story_data['title'] = TITLE
    story_data['level'] = LEVEL
    story_data['content'] = str_story_list[i]
    dict_list.append(story_data)
    
with open('../corpus/gutenberg_spanish_tales.json', 'w', encoding = 'utf-8') as outfile:
        json.dump(dict_list, outfile)
```

```python
# Ensure that we can read the data file 

with open('../corpus/gutenberg_spanish_tales.json', encoding = 'utf-8') as json_file:
    data = json.load(json_file)
    
print(type(data))
print(len(data))
print(type(data[0]))
print(data[0])
print(data[0]['content'])
```
