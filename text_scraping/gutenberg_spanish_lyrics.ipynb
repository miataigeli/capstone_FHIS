```python
#!pip install langid
from bs4 import BeautifulSoup
from urllib.request import urlopen
import langid
```

```python
###################################### Spanish Poems ###################################
#### stories stored as string in str_poem_list 


wb_address = 'https://www.gutenberg.org/files/16059/16059-h/16059-h.htm'
f = urlopen(wb_address)
html_soup = BeautifulSoup(f, 'lxml')
story_list = []
i = 0
for node in html_soup.find_all('div'):
    if node['class'] == ['poem']:
        i+= 1
        if langid.classify(node.get_text())[0] == 'es':
            story_list.append(node.get_text())
            
str_poem_list = []
for i in range(67, len(story_list)):
    str_poem_list.append(story_list[i])
```

```python
# Creating a json file

import json

SOURCE = wb_address
AUTHOR = 'ELIJAH CLARENCE HILLS, PH. D., LITT.D.'
TITLE = 'MODERN SPANISH LYRICS'
LEVEL = 'B'
dict_list = []

for i in range(len(str_poem_list)):
    story_data = {}
    story_data['source'] = SOURCE
    story_data['author'] = AUTHOR
    story_data['title'] = TITLE
    story_data['level'] = LEVEL
    story_data['content'] = str_poem_list[i]
    dict_list.append(story_data)
    
with open('gutenberg_spanish_lyrics.json', 'w', encoding = 'utf-8') as outfile:
        json.dump(dict_list, outfile)
```

```python
# Ensure that we can read the data file

with open('gutenberg_spanish_lyrics.json', encoding = 'utf-8') as json_file:
    data = json.load(json_file)
    
print(type(data))
print(len(data))
print(type(data[0]))
print(data[0])
print(data[0]['content'])
```
