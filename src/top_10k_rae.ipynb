##### http://corpus.rae.es/frec/10000_formas.TXT

```python
freq_dict = {}
with open("10000_formas.TXT", encoding="iso 8859-1") as f:
    for line in f.readlines()[1:]:
        num, word, abs_freq, _ = line.strip().split("\t")
        num = int(num[:-1])
        word = word.strip()
        abs_freq = int(abs_freq.strip().replace(",", ""))
        freq_dict[num] = [word, abs_freq]
```
