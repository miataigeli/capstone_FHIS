```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wiktionary.org/wiki/User:Matthias_Buchmeier/Spanish_frequency_list-1-5000"
```

```python
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
```

```python
body = soup.find("div", attrs={"class": "mw-parser-output"})
```

```python
freq_dict = {}
for i, f_w in enumerate(body.find("p").text.strip().split("\n")):
    freq_dict[i] = f_w.split(" ")[::-1]
```
